```python
### HTML-tegn
data_test['test_tekst'] = data_test['tekst'].str.replace(r'<[^<>]*>', '', regex=True)
### Backslashes
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'\\', '', regex=True)
# Fjerner {} klammer og indholdet imellem
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\sd\s\{\sfont.+;\s\}", "", regex=True)
### Ekstreme mellemrum
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'\s\s+', ' ', regex=True)
### Fjern mellemrum i starten
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'^\s', '', regex=True)

##### FJERNER TRÆLSE TEGN #### 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Ã¦', 'æ', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Ã¸', 'ø', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Ã¥', 'å', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Ã\x86', 'Æ', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Ã\x85', 'Å', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Â·', '', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'â\x80\x93', '', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'â\x80\x99s', '', regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r'Â ', '', regex=True) 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\s·", "", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\*\s", "", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\*","", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"Ã©", "é", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\s\\x96\s", "-", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\s\\x94\s", "\"", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"</p", "", regex=True)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"\s>", "", regex=True)


###### EKSTRA HTML #####
# Fjerner lange hyperlinks 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"<a\shref[^>].+>", "", regex=True)
#Fjerner ulukkede HTML-kodestykker (input)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"<input.+\sVerdana\"", "", regex=True)
#Fjerner ulukkede HTML-kodestykker (billeder)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"<img.+0\"", "", regex=True)
#Fjerner ulukkede HTML-kodestykker (tabeller)
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"<table.+535\"", "", regex=True)


##### SKABER TIL SIDST MELLEMRUM, DA TEKSTEN KAN VÆRE TRUKKET SAMMEN
# Skaber mellemrum mellem tal og ord, hvor der mangler
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?i)(?<=\d)(?=[a-z])|(?<=[a-z])(?=\d)", " ", regex=True)
# Skaber mellemrum mellem , og tal eller bogstaver 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=,)(?=\w)", " ", regex=True)
# Skaber mellemrum mellem : og tal eller bogstaver 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=[a-z|æ|ø|å]:)(?=\w)", " ", regex=True)
# Skaber mellemrum mellem . (med bogstaver foran) og store bogstaver
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=[a-z|æ|ø|å]\.)(?=[A-Z|Æ|Ø|Å])", " ", regex=True)
# Skaber mellemrum mellem . (med tal foran) og store bogstaver
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=[0-9]\.)(?=[A-Z|Æ|Ø|Å])", " ", regex=True)
# Skaber mellemrum mellem ) og bogstaver eller tal
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=\))(?=\w)", " ", regex=True)
# Skaber mellemrum mellem ord 
data_test['test_tekst'] = data_test['test_tekst'].str.replace(r"(?<=[a-z|æ|ø|å])(?=[A-Z|Æ|Ø|Å])", " ", regex=True)
```
